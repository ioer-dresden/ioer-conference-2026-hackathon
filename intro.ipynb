### Digitalisation meets Transformative Governance
**A Collaborative Data Challenge at the 4th IOER Conference "Space & Transformation"**

Welcome to the digital workspace of the `Conference Hack`. 

This Jupyter Book serves as the central hub for the `Sustainability Transformation Hackathon`, a contribution format (F5) of the [4th IOER Conference 2026](https://www.ioer.de/conference). Here, we bring together spatial data analysts, planners, coders, and domain experts to explore how Artificial Intelligence and Open Data can foster participation and sustainability.

### What is a "Conference Hack"?
Unlike traditional hackathons that require overnight coding marathons, this format is designed for `scientific collaboration and reproducibility`. 

We invite you to participate in an `asynchronous collaborative process` starting weeks before the conference. We provide a technical "Starter Kit"—including ready-to-use coding environments and curated IOER datasets—allowing you to develop your contributions at your own pace.

The goal is to co-create a `living publication`. All accepted contributions will be merged into this Jupyter Book, published as an `ioerDATA publication`, and assigned a DOI. This creates a lasting, citable artifact of our joint efforts.


### The Challenge Topics
We invite contributions that explore the intersection of `Digitalisation` and `Transformative Governance`. Specifically, we are looking for data-driven insights, workflows, or visualizations addressing:

*   **Urban Climate Resilience:** Heat adaptation, flood risk, and green infrastructure.
*   **Circularity:** Material flows, recycling potentials, and circular construction.
*   **Land Use Change:** Monitoring settlement dynamics and sprawl.
*   **Material Cadastres:** Mining the anthroposphere for future resources.


### Why Participate?
*   **Create Lasting Impact:** Your code and analysis will not disappear after the conference. It becomes part of a citable open science product (with DOI).
*   **Low Barrier to Entry:** You do not need to be a Python expert. We provide a "Starter Kit" featuring the `IOER Monitor API`, allowing you to pull and visualize spatial data immediately.
*   **Interdisciplinary:** We welcome contributions ranging from technical code (Python/R) to conceptual data stories or visualizations.
*   **Conference Pitch:** Selected teams will be invited to present their results in 5-minute pitches during a dedicated session at the main conference in Dresden (Sept 2026).


### How it Works
1.  **Register:** Sign up for the "Sustainability Transformation Hackathon" (Format F5) via the [Conference Registration Tool](https://www.ioer.de/conference).
2.  **Get the Kit:** Access the [Starter Kit](./workflow/01_starter_kit.ipynb), which includes tutorials on accessing IOER data via API and setting up your environment.
3.  **Collaborate:** Work on your idea asynchronously. You can fork this repository, add your notebook, and submit a Pull Request.
4.  **Publish:** We review and merge contributions into this book.
5.  **Present:** Join us in Dresden on September 23-24, 2026, to discuss the results.


### About the Infrastructure
This project is powered by the `IOER Research Data Centre (IOER-FDZ)`. We advocate for the `FAIR principles` (Findable, Accessible, Interoperable, Reusable). By using this Jupyter Book, your research is automatically reproducible, as it runs in the browser via [Binder](https://mybinder.org/) and [Jupyter4NFDI](https://jupyter.nfdi4biodiversity.org/).
